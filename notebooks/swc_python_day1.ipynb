```python
import numpy
import matplotlib.pyplot as plt

data = numpy.loadtxt('../data/chlorophyll-01.csv', delimiter = ',')

ave_profile = numpy.mean(data, axis=0)
std_profile = numpy.std(data, axis=0)
max_profile = numpy.max(data, axis=0)
min_profile = numpy.min(data, axis=0)

fig = plt.figure(figsize=(16,16))
ax1 = fig.add_subplot(221)
ax1.plot(ave_profile, depth)
ax1.set_xlabel('Chlorophyll Concentration [$\mu$g/m$^3$]', fontsize=14)
ax1.set_ylabel('Depth [m]', fontsize=14)
ax1.set_title(1)
ax2 = fig.add_subplot(222)
ax2.plot(std_profile, depth)
ax2.set_xlabel('Chlorophyll Concentration [$\mu$g/m$^3$] Standard Deviation', fontsize=14)
ax2.set_ylabel('Depth [m]', fontsize=14)
ax2.set_title(2)
ax3 = fig.add_subplot(223)
ax3.plot(max_profile, depth)
ax3.set_xlabel('Chlorophyll Concentration [$\mu$g/m$^3$] Maximum', fontsize=14)
ax3.set_ylabel('Depth [m]', fontsize=14)
ax3.set_title(3)
ax4 = fig.add_subplot(224)
ax4.plot(min_profile, depth)
ax4.set_xlabel('Chlorophyll Concentration [$\mu$g/m$^3$] Minimum', fontsize=14)
ax4.set_ylabel('Depth [m]', fontsize=14)
ax4.set_title(4)
```

```python
i=1

files = sorted(glob.glob('../data/chlorophyll-*.csv'))
print(files)

for f in files[1:13]:

    data = numpy.loadtxt(f, delimiter = ',')

    ave_profile = numpy.mean(data, axis=0)
    std_profile = numpy.std(data, axis=0)
    max_profile = numpy.max(data, axis=0)
    min_profile = numpy.min(data, axis=0)

    fig = plt.figure(figsize=(16,16))
    ax1 = fig.add_subplot(221)
    ax1.plot(ave_profile, depth)
    ax1.set_xlabel('Chlorophyll Concentration [$\mu$g/m$^3$]', fontsize=14)
    ax1.set_ylabel('Depth [m]', fontsize=14)
    ax1.set_title(1)
    ax2 = fig.add_subplot(222)
    ax2.plot(std_profile, depth)
    ax2.set_xlabel('Chlorophyll Concentration [$\mu$g/m$^3$] Standard Deviation', fontsize=14)
    ax2.set_ylabel('Depth [m]', fontsize=14)
    ax2.set_title(2)
    ax3 = fig.add_subplot(223)
    ax3.plot(max_profile, depth)
    ax3.set_xlabel('Chlorophyll Concentration [$\mu$g/m$^3$] Maximum', fontsize=14)
    ax3.set_ylabel('Depth [m]', fontsize=14)
    ax3.set_title(3)
    ax4 = fig.add_subplot(224)
    ax4.plot(min_profile, depth)
    ax4.set_xlabel('Chlorophyll Concentration [$\mu$g/m$^3$] Minimum', fontsize=14)
    ax4.set_ylabel('Depth [m]', fontsize=14)
    ax4.set_title(4)
    plt.title(f,fontsize=18)
    plt.savefig('../figs/chlorophyll_stats_%s.png' %i, dpi=150)
    i=i+1
```
